`np.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)`
- start: The starting value of the sequence.
- stop: The end value of the sequence.
- num (optional): The number of samples to generate. Default is 50.
- endpoint (optional): If True, stop is the last value in the range. If False, it is not included. Default is True.
- retstep (optional): If True, returns (samples, step), where step is the spacing between samples. Default is False.
- dtype (optional): The type of the output array. If None, the data type is inferred.
- axis (optional): The axis in the result along which the linspace values are stored. Default is 0.

###  A First Application: Classifying Iris Species
- length and width of the petals
- the length and width of the sepals.
 - species *setosa, versicolor, or virginica*.
- why supervised?
  -  Because we have measurements for which we know the correct species of iris, this is a
 supervised learning problem.
- predict one of several
 options (the species of iris). This is an example of a ***classification problem**
- The possible outputs (different species of irises) are called **classes**,this is a 3-classification problem.
- The desired output for a single data point (an iris) is the species of this flower. For a
 particular data point, the species it belongs to is called its **label**.

`train_test_split(*arrays, test_size=None, train_size=None, random_state=None, shuffle=True, stratify=None)`
- *** *arrays***:The input datasets to split. This can be one or more arrays or matrices (e.g., features and labels).
- ***test_size (optional)*** :The proportion of the dataset to include in the test split. This can be a float (representing a fraction), an integer (representing the absolute number of test samples), or None (automatically set to 0.25 if train_size is also None).
- ***train_size (optional)***:The proportion of the dataset to include in the train split. It works similarly to test_size. If None, it will be set to the complement of test_size.
- ***random_state (optional)***:Controls the shuffling applied to the data before splitting. Pass an integer for reproducible output across multiple function calls.`0`
  - we provide the pseudorandom number generator with a fixed seed using the
 random_state parameter.
- ***shuffle (optional)***:Whether or not to shuffle the data before splitting. Default is True.
- ***stratify (optional)***:If not None, the data is split in a stratified fashion, using this as the class labels. This helps maintain the same proportion of class labels in the train and test sets.

````python
from sklearn.model_selection import train_test_split
````

````python
X_train,X_test,y_train,y_test=train_test_split(iris['data'],iris['target'],random_state=0)
````

### scatter_matrix
- The `scatter_matrix `function in Pandas is a great way to visualize ***pairwise relationships*** between *features* in a dataset
- It's part of the `pandas.plotting` module and creates a grid of ***scatter plots*** for **every combination of numerical features** in a DataFrame.
- The **diagonal** of the grid is typically filled with ***histograms*** (or sometimes kernel density plots) of each individual feature.
- param
  - `alpha`: Transparency of points (0 to 1).
  - `figsize`: Size of the plot (width, height).
  - `diagonal`: 'hist' (histograms) or 'kde' (density plots).
  - `marker`: Marker style (e.g., 'o', 'x').
  - `hist_kwds`: Custom options for histograms, like bins.
  - `s` sets the size of the scatter points

````python
from pandas.plotting import scatter_matrix
grr=scatter_matrix(iris_df,c=y_train,figsize=(15,15),alpha=0.8,marker='o',hist_kwds={'bins':20},cmap='viridis')
````

### k-Nearest Neighbors
- The `k` in k-nearest neighbors signifies that instead of using only the closest neighbor
 to the new data point, we can consider any fixed number k of neighbors in the train
ing (for example, the closest three or five neighbors). Then, we can make a prediction
 using the majority class among these neighbors.
- An `Estimator` is a model or algorithm that learns from data. In scikit-learn, any model or algorithm that is used to fit data is considered an Estimator.
- All machine learning models in scikit-learn are implemented in their own classes,
 which are called `Estimator classes`.
- The *k-nearest neighbors* classification algorithm
 is implemented in the `KNeighborsClassifier` class in the `neighbors` module.
 > Before
 we can use the model, ***we need to instantiate the class into an object***.
- This is when we
 will set any parameters of the model. The most important parameter of KNeighbor
 sClassifier is the *number of* `neighbors`.



````python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
````

>The `knn object` encapsulates the algorithm that will be used to build the model from
 the training data, as well the algorithm to make predictions on new data points. It will
 also hold the information that the algorithm has extracted from the training data. In
 the case of KNeighborsClassifier, it will just store the training set.

---
### => fit()
- *Inputs to fit()*:
  - `X`: The training data (usually a 2D array or DataFrame) representing the features (input variables).
  - `y`: The target labels (output variables) corresponding to the training data X. For supervised learning tasks, this is the "ground truth" data the model aims to predict.

- *What Happens During fit()*:

  - ***Learning Process***: The algorithm examines the training data (X_train) and learns the relationship between the features and the target values (y_train).
  - ***Model Parameters***: The algorithm updates its internal parameters to best map the inputs (features) to the outputs (targets). This could involve finding coefficients, weights, thresholds, or distances, depending on the type of model you're using.
  
  > **Note** that we made the measurements of this single flower into a row in a `two
dimensional NumPy array`, as scikit-learn always expects two-dimensional arrays
 for the data


````python
knn.fit(X_train,y_train)
````

````python
X_new=np.array([[5,2.9,1,0.2]])
prediction=knn.predict(X_new)
print(prediction,iris['target_names'][prediction])
````

````python
y_pred=knn.predict(X_test)
print(y_pred)
````

````python
print(np.mean(y_pred==y_test))
print(knn.score(X_test,y_test))#n_neighbor=1,5 yields the same result
````
